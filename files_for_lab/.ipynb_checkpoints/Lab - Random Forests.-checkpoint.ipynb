```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
pd.set_option('display.max_columns',None)
categorical = pd.read_csv('categorical.csv')
numerical = pd.read_csv('numerical.csv')
target = pd.read_csv('target.csv')
```

```python
X = pd.concat([categorical, numerical], axis=1)
y = target['TARGET_B']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop='first')
X_train_cat_encoded = enc.fit_transform(X_train_cat).toarray()
X_test_cat_encoded = enc.transform(X_test_cat).toarray()
```

```python
X_train = pd.concat([pd.DataFrame(X_train_num_scaled), pd.DataFrame(X_train_cat_encoded)], axis=1)
X_test = pd.concat([pd.DataFrame(X_test_num_scaled), pd.DataFrame(X_test_cat_encoded)], axis=1)
```

# Managing imbalance in the dataset

```python
y_train = y_train.reset_index().drop(['index'], axis=1)
```

```python
trainset= pd.concat([X_train, y_train], axis=1)
trainset
```

```python
from sklearn.utils import resample

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

## Apply the Random Forests algorithm but this time only by upscaling the data.

```python
category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
X_train_upsampled = data_upsampled.drop(['TARGET_B'], axis =1)
y_train_upsampled = data_upsampled['TARGET_B'] 
```

```python
from sklearn.ensemble import RandomForestClassifier
# Fit the Random Forests algorithm on the training data
rfc = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
rfc.fit(X_train_upsampled, y_train_upsampled)
print(rfc.score(X_train_upsampled,y_train_upsampled))
print(rfc.score(X_test,y_test))
# Make predictions on the test data
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

# Calculate the confusion matrix of the model
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion matrix:\n', conf_matrix)

# For cross validation
cross_val_scores = cross_val_score(rfc, X_train_upsampled, y_train_upsampled, cv=10)
print(np.mean(cross_val_scores))
```

## Use Feature Selections that you have learned in class to decide if you want to use all of the features (Variance Threshold, RFE, PCA, etc.)

```python
from sklearn.feature_selection import VarianceThreshold, RFE
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
X_added_constant = sm.add_constant(X_train_upsampled)
X_added_constant
model = sm.OLS(y_train_upsampled,X_added_constant).fit()
model.summary()
# Drop columns with pvalue >= 0.8 -> 26,27,43,99,115,174,178,193,205,218,245,271,294,322
```

```python
X_train_upsampled = X_train_upsampled.drop([26,27,43,99,115,174,178,193,205,218,245,271,294,322], axis=1)
X_test = X_test.drop([26,27,43,99,115,174,178,193,205,218,245,271,294,322], axis=1)
```

```python
# Fit the Random Forests algorithm on the training data
rfc = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
rfc.fit(X_train_upsampled, y_train_upsampled)
print('Accuracy Train sample with Random Forest: 'rfc.score(X_train_upsampled,y_train_upsampled))
print('Accuracy Test sample with Random Forest: 'rfc.score(X_test,y_test))
# Make predictions on the test data
y_pred = rfc.predict(X_test)
```

We found improvement by reducing the number of features which means that there are still irrelevant features. For this reason, we use different feature selection tools below

```python
# Use variance threshold to remove low-variance features
vt = VarianceThreshold(threshold=0.05)
X_vt = vt.fit_transform(X_train_upsampled)

# Use PCA to reduce the dimensionality of the data
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_train_upsampled)

# Model performance using all features
scores_all_features = cross_val_score(rfc, X_train_upsampled, y_train_upsampled, cv=5, scoring='accuracy')
print('Accuracy using all features:', scores_all_features.mean())

# Model performance using variance threshold
scores_vt = cross_val_score(rfc, X_vt, y_train_upsampled, cv=5, scoring='accuracy')
print('Accuracy using variance threshold:', scores_vt.mean())

# Model performance using PCA
scores_pca = cross_val_score(rfc, X_pca, y_train_upsampled, cv=5, scoring='accuracy')
print('Accuracy using PCA:', scores_pca.mean())
```
